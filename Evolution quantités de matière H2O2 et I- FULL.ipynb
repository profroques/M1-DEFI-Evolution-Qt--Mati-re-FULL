```python
import matplotlib.pyplot as plt #importation du module pyplot de bibliothèque matplotlib sous forme d'alias plt

#déclaration des nombres stoechiométriques des réactifs
a = 1 #coefficient pour l'eau oxygénée
b = 2 #coefficient pour les ions iodures
c = 1 #coefficient pour le diiode


#déclaration des quantités de matière initiales en mol
n0_H2O2 = 3.0E-3
n0_iodure = 2.5E-3
n0_diiode = 0

#création de liste associées aux quantités de matière en mol
liste_H2O2 = [n0_H2O2]
liste_iodure = [n0_iodure]
liste_diiode = [n0_diiode]

#à propos de l'avancement

x=[0] #Initialisation de l'avancement en mol
dx = 5E-5 #incrément de l'avancement en mol

'''
Script pour stocker dans les listes les valeurs des qtés de matière
de H2O2, I-, I2 au fur et à mesure que x augmente par pas de dx
'''
i = 0 #compteur d'itération

while (liste_H2O2[-1]>0) and (liste_iodure[-1]>0):
	x.append(x[i]+dx)
	liste_H2O2.append(liste_H2O2[i]-a*dx)
	liste_iodure.append(liste_iodure[i]-b*dx)
	liste_diiode.append(liste_diiode[i]+c*dx)
	i=i+1

'''
Fin du script
'''

#détermination de la valeur de l'avancemement maximal et bilan de matière à l'état final
xmax=x[-2]
nf_H2O2 = liste_H2O2[-2]
nf_iodure = liste_iodure[-2]
nf_diiode = liste_diiode[-2]

"""
print()
print("x : ",x)
print()
print("H2O2 : ",liste_H2O2)
print()
print("iodure : ", liste_iodure)
print()
print("diiode : ", liste_diiode)
print()
"""


#Affichage des valeurs de xmax et des quantités de matières à l'état final
print()
print("L''avancement maximal xmax vaut : xmax = ",format(xmax,'0.2E')," mol")
print("Quantité de matière de H2O2 à l'état finale nf_H2O2 = ",format(nf_H2O2,'.2E')," mol")
print("Quantité de matière d''ions iodure à l'état finale nf_iodure = ",format(nf_iodure,'.2E')," mol")
print("Quantité de matière de diiode à l'état finale nf_diiode = ",format(nf_diiode,'.2E')," mol")



#Affichage de la représentation graphique de l'évolution des quantités de matière en fonction de x
plt.plot(x,liste_H2O2, 'r-', label = "H2O2") #affichage de la qté de H2O2 en fonction de x avec droite trait plein en rouge
plt.plot(x,liste_iodure, 'g--', label = "I-") #affichage de la qté de I- en fonction de x avec droite pointillé en vert
plt.plot(x,liste_diiode, 'b*',label = "I2") #affichage de la qté de I2 en fonction de x avec droite en étoiles en bleu
plt.title("Evolution des quantités de matières en fonction de x") #Titre du graphique
plt.xlabel("avancement x (mol)") #titre de l'axe des abscisses
plt.ylabel("qté de matière (mol)") #titre de l'axe des ordonnées
plt.legend() #afficher la légende des courbes
plt.grid() #afficher une grille
plt.show() #afficher le graphique
```
